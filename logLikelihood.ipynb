```python
from collections import defaultdict
def processFile(filename):
    d = defaultdict(int)
    total = 0
    with open(filename) as file:
        for line in file:
            line = line.replace('/', ' / ')
            words = line.split()
            for word in words:
                word = word.strip("()'.,:-/[]\"")
                if len(word) > 0:
                    d[word] += 1
                    total += 1
    return(d, total)
            
processFile("resumes/female/female01.txt")
```

```python
cutoff = 5

def logLikelihood(filename1, filename2):
    d = defaultdict(int)
    (d1, tot1) = processFile(filename1)
    (d2, tot2) = processFile(filename2)
    for (word, count1) in d1.items():
        count2 = d2[word]
        if ((count1 > cutoff) or (count2 > cutoff)):
            ll = loglike(count1, tot1 - count1, count2, tot2 - count2)
            d[word] = [ll, word, count1, count2]
    for (word, count2) in d2.items():
        if word not in d:
            count1 = d1[word]
            if ((count1 > cutoff) or (count2 > cutoff)):
                ll = loglike(count1, tot1 - count1, count2, tot2 - count2)
                d[word] = [ll, word, count1, count2]
    results = list(d.values())
    results.sort(reverse=True)
    with open("output.txt", 'w') as file:
        for result in results:
            file.write("%7.4f  %20s  %5i  %5i\n" % (result[0], result[1], result[2], result[3] ))
    return results
        
        
logLikelihood("nurseJobs.txt", "developerJobs.txt")    
```
